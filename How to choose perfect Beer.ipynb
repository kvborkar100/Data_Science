```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#get dataset
train = pd.read_csv('Beer Train Data Set.csv')
test = pd.read_csv('Beer Test Data Set.csv')
```

```python
#combine train and test data
train['source'] = 'train'
test['source'] = 'test'
dataset = pd.concat([train,test],ignore_index=True)
print(train.shape,test.shape,dataset.shape)
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
# checking null values
dataset.isnull().sum()
```

```python
# number of unique values per column
dataset.apply(lambda x: x.nunique())
```

```python
dataset.dtypes
```

```python
from sklearn.preprocessing import Imputer
imputer = Imputer()
dataset.iloc[:,0:1]= imputer.fit_transform(dataset.iloc[:,0:1])
```

```python
dataset.isnull().sum()
```

```python
print(dataset['Cellar Temperature'].value_counts())
print(dataset['Serving Temperature'].value_counts())
```

```python
dataset['Cellar Temperature'] = dataset['Cellar Temperature'].replace(np.nan,'40-45')
```

```python
dataset['Serving Temperature'] = dataset['Serving Temperature'].replace(np.nan,'45-50')
```

```python
dataset.isnull().sum()
```

```python
print(dataset['Cellar Temperature'].value_counts())
print(dataset['Serving Temperature'].value_counts())
```

```python
dataset.dtypes
```

```python
dataset['Food Paring'] = dataset['Food Paring'].apply(lambda x: x.replace(')',','))
```

```python
dataset['Food Paring'][0]
```

```python
dataset['Food Paring'] = dataset['Food Paring'].apply(lambda x: len(x.split(',')))
```

```python
dataset.dtypes
```

```python
dataset['Glassware Used'] = dataset['Glassware Used'].apply(lambda x:len(x.split(',')))
```

```python
dataset.dtypes
```

```python
dataset['Style Name'].apply(lambda x: x if 'American' in x)
```

```python
def removec():
    for x in range(len(dataset['Ratings'])):
                   if ',' in dataset['Ratings'][x]:
                       dataset['Ratings'][x] = dataset['Ratings'][x].replace(',','')
removec()
```

```python
dataset['Ratings'] = dataset['Ratings'].apply(lambda x : int(x))
```

```python
dataset['Ratings']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
var = ['Cellar Temperature','Serving Temperature']
for i in var:
    le.fit_transform(dataset[i])
dataset= pd.get_dummies(dataset,columns=var)
```

```python
dataset.dtypes
```

```python
dataset.drop(columns=['Brewing Company','Beer Name','Style Name'],inplace=True)
```

```python
dataset
```

```python
train_mod = dataset.loc[dataset['source'] == 'train']
test_mod = dataset.loc[dataset['source'] == 'test']
```

```python
train_mod
```

```python
test_mod.shape
```

```python
train_mod.drop(columns=['source'],inplace=True)
```

```python
test_mod.drop(columns=['source','Score'],inplace=True)
```

```python
train_mod.columns
```

```python
train_df1 = train_mod.iloc[:,0:4]
```

```python
train_df2 = train_mod.iloc[:,5:]
```

```python
demo_train = pd.concat([train_df1,train_df2], axis=1,join_axes=[train_df1.index])
```

```python
demo_test = train_mod['Score']
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression(normalize=True)
regressor.fit(demo_train,demo_test)
```

```python
y_pred = regressor.predict(test_mod)
```

```python
output = pd.DataFrame(y_pred,columns=['Score'])
```

```python
output.to_excel('Output.xlsx',index=False)
```

```python
from sklearn.model_selection import cross_val_score
accuracy = cross_val_score(estimator = regressor,X= demo_train,y=demo_test,cv = 5)
```

```python
accuracy
```

```python
accuracy.std()
```
